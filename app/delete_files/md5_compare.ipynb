# This module aims to compare files in order to find duplicates

```python
import hashlib
import os

import customtkinter as ctk
from customtkinter import filedialog
import pandas as pd

# Customtk general settings
ctk.set_appearance_mode("dark")

pot_files = "../../resources/potential_first_check_py_files"

class FileHandling():
    def __init__(self):
        pass

    def selecfiles(self):
        filenames = filedialog.askopenfilenames(initialdir=pot_files)
        # print(filenames)
        return filenames
    
    # def is_this_visible(self):
    #     print(self.filenames)

```

```python
gottatrythis = FileHandling()

md5_list = {}

for file in gottatrythis.selecfiles():
    with open(file, 'rb') as myfile:
        # after open the file has to be read
        myfile = myfile.read()
        # retrieve the file content md5 equivalent
        md5 = hashlib.md5(myfile).hexdigest()

        # get file size in bytes
        filesize = os.path.getsize(file)

        # get filename
        filename = os.path.basename(file)
        
        # put together in a dictionary
        md5_list.update({filename : {"hashed_md5" : md5, "path": file, "size" : filesize}})

# print(md5_list)

df = pd.DataFrame.from_dict(md5_list, orient="index")
display(df)        
```

```python
# Finds all files that has a duplicate
df.duplicated(subset=['hashed_md5'])

# create delete-list
df_delete_list = df[df.duplicated(subset=['hashed_md5'])]

print(f"df-count {df.count()}") #19
print(f"df_delete_list-count {df_delete_list.count()}") #13

# display(df_delete_list)
```

```python
print(df_delete_list.path)
```

```python
# this works! :D
for path in df_delete_list['path']:
    try:
        os.remove(path)
        print(f"Deleted: {path}")
    except FileNotFoundError:
        print(f"File not found: {path}")
    except Exception as e:
        print(f"Error deleting {path}: {e}")
```
